```python
#API url to get initial list of movies in dataset
bechdel_url = 'http://bechdeltest.com/api/v1/getAllMovies'
```

```python
#create dataframe from initial API request
bechdel_all = pd.DataFrame(raw_data,columns = ['id','imdbid','rating','title','year'])

bechdel_all
```

```python
#set up API to gather bechdel testing data

# ****WARNING******  This took appx 3 hours to run - Run at your own risk.

bechdel_url2 = 'http://bechdeltest.com/api/v1/getMovieByImdbId?imdbid='
movies = bechdel_all['imdbid']

visible = []
date_added = []
bechdel_rating = []
imdbid = []
bechdel_id = []
title = []
release_year = []

for movie in movies:
    bechdel_query = bechdel_url2 + movie
    bechdel_response = requests.get(bechdel_query)
    raw_data2 = bechdel_response.json()
    try:
        visible.append(raw_data2['visible'])
        date_added.append(raw_data2['date'])
        bechdel_rating.append(raw_data2['rating'])
        imdbid.append(raw_data2['imdbid'])
        bechdel_id.append(raw_data2['id'])
        title.append(raw_data2['title'])
        release_year.append(raw_data2['year'])
    except:
        continue
```

```python
bechdel_dict={'visible': visible,
             'date_added': date_added,
             'bechdel_rating': bechdel_rating,
             'imdb_id': imdbid,
             'bechdel_id': bechdel_id,
             'title': title,
             'release_year': release_year
             }
bechdel_df=pd.DataFrame(bechdel_dict)
bechdel_df
```
